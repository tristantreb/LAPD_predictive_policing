```python
def group_crimes(crime):
    if(crime in assault):
        return 'Assault'
    if(crime in Theft):
        return 'Theft'
    if(crime in Robbery):
        return 'Robbery'
    if(crime in Burglary):
        return 'Burglary'
    if(crime in Motor_Vehicle):
        return 'Motor_Vehicle'
    if(crime in Arson):
        return 'Arson'
    if(crime in Homicide):
        return 'Homicide'
    if(crime in Vice):
        return 'Vice'
    if(crime in Violence):
        return 'Violence'
    if(crime in Others):
        return 'Others'
    return crime

data_processed['Crime Code Description'] = data_processed['Crime Code Description'].apply(lambda c: group_crimes(c))

```

```python
 #there is 2 dataset for dangerous and not dangerous
new_df1 = pd.read_csv(DATA_ROOT+"new_df_y_2016_dangerous.csv")  #deadly model
##new_df1 = pd.read_csv(DATA_ROOT+"new_df_y_2016_not_dangerous.csv")  #perilous model

new_df1 
```

```python
#new_df1=new_df.copy()
new_df1['y'] = new_df1['y'].map(lambda s: 0 if s==0 else 1)
```

```python
burglary_count=new_df1['y'].groupby(new_df1['y']).count().sort_values(0, ascending=False)
```

```python
new_df1.y.value_counts().plot(kind='pie')
```

```python
Y = new_df1['y']
df = new_df1.drop(columns="y")


X_train, X_test, y_train, y_test = train_test_split(df, Y, test_size=0.15, random_state=1) # 80% training and 20% test


ros = RandomOverSampler(random_state=0, sampling_strategy='minority')
X_train, y_train = ros.fit_resample(X_train, y_train)
y_train.value_counts().plot(kind='pie')

```

**ML Function**

```python
def accuracy(TP, TN, FP, FN):
    return (TP + TN) / (TP + TN + FP + FN)

def recall(TP, TN, FP, FN):
    return TP/(TP + FN)

def precision(TP, TN, FP, FN):
    return TP/(TP + FP)
```

**Decision Tree Method**

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="gini", max_depth=5)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

([tn, fp], [fn, tp]) = sk.metrics.confusion_matrix(y_test, y_pred)

R=recall(tp,tn,fp,fn)

P=precision(tp,tn,fp,fn)



print(R)
print(P)
```

**KNN Method**

```python
neigh = sk.neighbors.KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_train, y_train)
y_pred_knn = neigh.predict(X_test)
([tn_knn, fp_knn], [fn_knn, tp_knn]) = sk.metrics.confusion_matrix(y_test, y_pred_knn)
acc_knn = accuracy(tp_knn, tn_knn, fp_knn, fn_knn)


R_knn=recall(tp_knn,tn_knn,fp_knn,fn_knn)
P_knn=precision(tp_knn,tn_knn,fp_knn,fn_knn)
print (acc_knn)
print(R_knn)
print(P_knn)
```

```python
numbers = list(range(1,30))
knn_scores = []
for k in numbers:
    knn = sk.neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)
    ([tn, fp], [fn, tp]) = sk.metrics.confusion_matrix(y_test, y_pred_knn)
    acc_knn = accuracy(tp, tn, fp, fn)
    knn_scores.append(acc_knn)
  



best_k = numbers[knn_scores.index(max(knn_scores))]
print ('best k = %d with accuracy score = %2.4f' %(best_k, max(knn_scores)))

plt.plot(numbers,knn_scores)
```

**Logistic Regression Method**

```python
y_train_logreg=y_train.astype('int')
y_test_logreg = y_test.astype('int')

logreg = LogisticRegression()
logreg.fit(X_train, y_train_logreg)

y_pred_log = logreg.predict(X_test)

([tn_log, fp_log], [fn_log, tp_log]) = sk.metrics.confusion_matrix(y_test, y_pred_log)
R_log=recall(tp_log,tn_log,fp_log,fn_log)
P_log=precision(tp_log,tn_log,fp_log,fn_log)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test_logreg)))
print(R_log)
print(P_log)
```
